## Réalisation d’une carte Cloropleth

### Confirmation de votre savoir-faire :

#### - A partir d’une source de données que vous aurez glané sur Internet
    
#### - Pour une région d’intérêt que vous présenterez réalisez une carte Cloropleth en utilisant pandas. 

```python
import geopandas
import folium
import pandas
import json
import os

# afficher le fond dr carte osm
paris = folium.Map(location = [48.856578, 2.351828], zoom_start = 12)

# Chargement du fichier geographique
geo = json.load(open("arrondissements.geojson"))

# affichage de la géometrie de notre couche
paris.choropleth(geo_data = geo)

#Création du DataFrame
df = pandas.DataFrame({
    "Arrondissement" : [arr["properties"]["c_ar"] for arr in geo["features"]], 
    "Surface" : [arr["properties"]["surface"] for arr in geo["features"]]
})

# Carte Choropleth
paris = folium.Map(location = [48.856578, 2.351828], zoom_start = 12)
paris.choropleth(geo_data = geo, key_on = "feature.properties.c_ar",
                 data = df, columns = ["Arrondissement", "Surface"],
                 fill_color = "PRGn")
paris

#paris.save(os.path.join('carte_choropleth_0.html'))
```
